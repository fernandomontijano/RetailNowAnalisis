```python
# Paso 1: Importar las librerías necesarias
import os
import pandas as pd
import numpy as np
```

```python
# Paso 2: Cargar los datos desde los archivos CSV
# Se usan rutas relativas porque los CSV están en la misma carpeta del notebook
print(os.getcwd())
print(os.listdir())

ventas = pd.read_csv('ventas.csv')
inventarios = pd.read_csv('inventarios.csv')
satisfaccion = pd.read_csv('satisfaccion.csv')

ventas.head(), inventarios.head(), satisfaccion.head()
```

```python
# Paso 3: Limpieza de datos
# Eliminar filas con valores nulos
ventas = ventas.dropna()
inventarios = inventarios.dropna()
satisfaccion = satisfaccion.dropna()

# Comprobar el resultado de eliminar filas con valores nulos
print("No hay valores NULL en ventas:", "OK" if ventas.isnull().sum().sum() == 0 else "Hay nulos")
print("No hay valores NULL en inventarios:", "OK" if inventarios.isnull().sum().sum() == 0 else "Hay nulos")
print("No hay valores NULL en satisfaccion:", "OK" if satisfaccion.isnull().sum().sum() == 0 else "Hay nulos")
```

```python
# Paso 4: Exploración de datos (ventas)
# Ventas totales por producto y por tienda
print(ventas)

# Creamos el total de cada línea de venta (cantidad x precio unitario)
ventas["Ingresos"] = ventas["Cantidad_Vendida"] * ventas["Precio_Unitario"]

# Agrupamos por tienda y producto, sumando la cantidad vendida y el total vendido
ventas_totales_producto_tienda = (
    ventas.groupby(["ID_Tienda", "Producto"])
    .agg(
        Total_Cantidad_Vendida=("Cantidad_Vendida", "sum"),
        Total_Ingresos=("Ingresos", "sum")
    )
    .reset_index()
)

print("Ventas totales por producto en cada tienda:")
print(ventas_totales_producto_tienda)
```

```python
# Ventas totales por tienda
print(ventas)

# Agrupamos por tienda y sumamos el total de venta
total_venta_tienda = (
    ventas.groupby("ID_Tienda")["Ingresos"]
    .sum()
    .reset_index()
)
print("Total de venta por tienda:")
print(total_venta_tienda)

total_cantidad_vendida_tienda = (
    ventas.groupby("ID_Tienda")["Cantidad_Vendida"]
    .sum()
    .reset_index()
)
print("Cantidad total de productos vendidos por tienda:")
print(total_cantidad_vendida_tienda)
```

```python
# Paso 4 (cont.): Ingresos y resumen estadístico
# Ingresos totales por tienda (cantidad * precio)
#Realizado en el paso anterior, pero por si acaso vuelvo hacerlo por validación de la tarea
print(total_venta_tienda)

ingresos_totales_tienda = (
    ventas.groupby("ID_Tienda")["Ingresos"]
    .sum()
    .reset_index()
)
print("Total de ingresos por tienda:")
print(ingresos_totales_tienda)
```

```python
# Resumen estadistico de ventas
print("Resumen estadístico de las ventas:")
# describe() nos da media, desviación estándar, mínimo, máximo y percentiles
print(ventas["Ingresos"].describe())
# La mediana se corresponde con el percentil 50%, pero por si acaso la muestro individual
print("Mediana de las ventas:", ventas["Ingresos"].median())
print("Media de las ventas:", ventas["Ingresos"].mean())
```

```python
# Promedio de ventas por tienda y producto
# Usamos mean() en lugar de sum() para obtener el valor medio de cada combinacion
print(ventas.groupby(["ID_Tienda", "Producto"]).size())
promedio_ventas_tienda_producto = (
    ventas.groupby(["ID_Tienda", "Producto"])["Ingresos"]
    .mean()
    .reset_index()
)

# Como cada producto se vende una sola vez por tienda, el promedio coincide con el total
print("Promedio de ventas por tienda y producto:")
print(promedio_ventas_tienda_producto)
```

```python
# Paso 5: Analisis de inventarios
# Stock total por tienda

# Calculamos unidades vendidad por tienda y producto
unidades_vendidas = (
    ventas.groupby(["ID_Tienda", "Producto"])["Cantidad_Vendida"]
    .sum()
    .reset_index()
)

# Revisamos con el inventario las unidades vendidas (Hay que tener en cuenta tienda y producto)
inventarios = inventarios.merge(
    unidades_vendidas, on=["ID_Tienda", "Producto"], how="left"
)
print(inventarios)

# Revisar valores NaN e inventarios, en caso de estar en NaN ponerlo a 0 para su correcto uso
inventarios["Cantidad_Vendida"] = inventarios["Cantidad_Vendida"].fillna(0)
```

```python
# Rotacion de inventarios por tienda (ventas totales / stock total)
inventarios["Rotacion"] = (
    inventarios["Cantidad_Vendida"] / inventarios["Stock_Disponible"]
)
print("Rotacion de inventarios por tienda:")
print(inventarios)
```

```python
# Tiendas con inventario critico (< 10% de ventas respecto al stock)
inventario_critico = inventarios[inventarios["Rotacion"] < 0.10]
# Versión en porcentaje, por claridad en el informe
print("Inventario crítico (se vendió menos del 10% del stock disponible):")
print(inventario_critico)

# Resultado vacio, no se han encontrado productos que se han vendido menos del 10%. Como el resultado no me gusta simplemente voy a mejorarlo de esta forma
# Para mejorar la presentacion
if inventario_critico.empty:
    print("No se han detectado tiendas con inventario crítico, todas superan el 10% de rotación.")
else:
    print("Inventario crítico, se vendió menos del 10% del stock disponible:")
    print(inventario_critico)

# Rotación media por tienda: EXTRA
rotacion_por_tienda = inventarios.groupby("ID_Tienda")["Rotacion"].mean().reset_index()
print("Rotación media por tienda:")
print(rotacion_por_tienda)
```

```python
# Paso 6: Satisfaccion del cliente
# Unir ventas totales con satisfaccion para relacionar rendimiento
satisfaccion_ventas = satisfaccion.merge(
    ingresos_totales_tienda, on="ID_Tienda", how="left"
)

print("satisfaccion y ventas totales por tienda:")
print(satisfaccion_ventas)
```

```python
# Tiendas con satisfaccion menor al 60%
menor_satisfaccion = satisfaccion_ventas[satisfaccion_ventas["Satisfacción_Promedio"] < 60]
if menor_satisfaccion.empty:
    print("No hay tiendas con satisfaccion por debajo del 60%.")
else:
    print("Tiendas con satisfaccion menor al 60% (con sus ventas):")
    print(menor_satisfaccion)
```

```python
# Paso 7: Operaciones con Pandas (mediana y desviacion estandar)
mediana_ventas = ingresos_totales_tienda["Ingresos"].median()
desviacion_ventas = ingresos_totales_tienda["Ingresos"].std()

print("Mediana de las ventas totales por tienda:", mediana_ventas)
print("Desviacion estandar de las ventas totales:", desviacion_ventas)
```
